```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from grover import *
```

```python
############ String to be found '111' using noancilla type ############
test_1, noise_model, basis_gates = grover(['111'], 'noancilla',0.001, 0.01)
test_1.decompose().draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
count1 = execute(test_1,simulator,basis_gates=basis_gates,
                 noise_model=noise_model).result().get_counts()
print(count1)
plot_histogram(count1,color='limegreen')
```
