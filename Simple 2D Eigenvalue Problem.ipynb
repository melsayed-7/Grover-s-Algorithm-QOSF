```python
ListExpectationValue=[]
theta1=[] #x-axis
step=pi/32
for n in range(100):
    circuit1=QuantumCircuit(1,1)
    circuit1.ry(n*pi/32,0)
    theta=n*pi/32
    circuit1.ry(-pi/2,0)
    circuit1.measure(0,0)
    simulator=Aer.get_backend('qasm_simulator')
    result=execute(circuit1,simulator).result()
    measurement=result.get_counts()
    ExpectationValue=(((measurement.get('0',0) -measurement.get('1',0))/1024)-1) #the -1 is for the identity matrix
    ListExpectationValue.append(ExpectationValue)
    theta1.append(theta)
plt.scatter(theta1,ListExpectationValue) 
plt.xlabel('theta') 

plt.ylabel('Expectation Value') 
plt.show() 
```
